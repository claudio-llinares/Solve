```python
import multiprocessing as mp
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
#path="../../solve_runs/runs/run_solve_ics_128_512mpc/"
path="../../solve_runs/runs/run_solve_ics_064_064mpc/"
box  = 64.0
grid = 64

rows = np.arange(grid)
cols = np.arange(grid)
coords = np.empty((grid, grid, 2), dtype=np.intp)
coords[..., 0] = rows[:, None]
coords[..., 1] = cols

# make one plot:
def plot_one(plot_number):
    print plot_number, "%s/solve.phi.%05i" % (path, plot_number+1)
    
    data=open("%s/solve.phi.%05i" % (path, plot_number+1))
    tmp1 = np.fromfile(data, dtype=np.float64, count = 4*grid**2)
    data.close()
    # Reshape density
    tmp2 = np.reshape(tmp1, [grid**2,4])
    phi     = np.reshape(tmp2[:,1], [grid, grid])
    phi_dot = np.reshape(tmp2[:,2], [grid, grid])
    lap_phi = np.reshape(tmp2[:,3], [grid, grid])
    rho     = np.reshape(tmp2[:,0], [grid, grid])
    
    plt.figure(figsize=(10,7)) #40, 35)) #15))

    plt.subplot(2,2,1)
    plt.grid()
    xx = np.linspace(0, box, grid) # WARNING: there is no shift in here.
    min = np.amin(phi)
    max = np.amax(phi)
    #plt.ylim(-1.0, 1.0)
    #plt.autoscale(enable=True)
    plt.pcolormesh(xx, xx, phi, vmin=min, vmax=max)  #, cmap='inferno')
    #plt.plot(xx, phi)
    plt.colorbar()
        
    plt.subplot(2,2,2)
    plt.grid()
    min = np.amin(phi_dot)
    max = np.amax(phi_dot)
    #plt.ylim(-1.0, 1.0)
    #plt.autoscale(enable=True)
    plt.pcolormesh(xx, xx, phi_dot, vmin=min, vmax=max)  #, cmap='inferno')
    #plt.plot(xx, phi_dot)    
    plt.colorbar()
    
    plt.subplot(2,2,3)
    plt.grid()
    min = np.amin(lap_phi)
    max = np.amax(lap_phi)
    #plt.ylim(-0.005, 0.005)
    #plt.autoscale(enable=True)
    plt.pcolormesh(xx, xx, lap_phi, vmin=min, vmax=max)  #, cmap='inferno')
    #plt.plot(xx, lap_phi)
    plt.colorbar()
    
    plt.subplot(2,2,4)
    plt.grid()
    min = np.amin(rho)
    max = np.amax(rho)
    #plt.ylim(-0.005, 0.005)
    #plt.autoscale(enable=True)
    plt.pcolormesh(xx, xx, rho, vmin=min, vmax=max)  #, cmap='inferno')
    #plt.plot(xx, lap_phi)
    plt.colorbar()
    

    plt.savefig('map_%05i' % (plot_number + 1))
    file='hola_%05i.dat' % (plot_number + 1)
    #print coords
    print np.reshape(coords, [grid**2])
    np.savetxt(file, np.column_stack((np.reshape(coords, [1,grid**2]), np.reshape(phi, [1,grid**2])))) #, fmt="%g  %g")
```

```python
# Count number of processes:
pool = mp.Pool(mp.cpu_count())

# Make all the plots in parallel:
plots = xrange(8)
output_list = pool.map(plot_one, plots) 
```
